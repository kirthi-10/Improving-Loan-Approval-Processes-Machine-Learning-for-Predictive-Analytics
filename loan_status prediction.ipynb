


*   By using standard scaling we can achieve maximum accuracy for our model.

```python
x=df.drop(['Loan_Status'],axis=1)
y=df[['Loan_Status']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
def correlation (df, threshold):

   col_corr = set() # Set of all the names of correlated columns corr_matrix dataset.corr()
   corr_matrix=df.corr()
   for i in range(len(corr_matrix.columns)):

        for j in range(1):

          if (corr_matrix.iloc[i, j]) > threshold:
            colname =corr_matrix.columns[i] # getting the name of column col_corr.add(colname)
            col_corr.add(colname)
            return col_corr
```

```python
corr_features=correlation(x_train,0.4)
len(set(corr_features))
corr_features
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred1=lr.predict(x_test)
```

```python
accuracy1=accuracy_score(y_test,y_pred1)
accuracy1
```

```python
cm_test = confusion_matrix(y_test, y_pred1)
matrix = ConfusionMatrixDisplay(confusion_matrix = cm_test, display_labels = [0, 1])
matrix.plot(cmap="Blues")
plt.title("ConfusionMatrix.", weight="bold");
```

```python
cr1=classification_report(y_test,y_pred1)
print(cr1)
```

# **DecisionTree**

```python
from sklearn.tree import DecisionTreeClassifier
dr=DecisionTreeClassifier(max_depth=5,criterion='gini',max_features=5,max_leaf_nodes=5)
dr.fit(x_train,y_train)
y_pred2=dr.predict(x_test)
```

```python
accuracy2=accuracy_score(y_test,y_pred2)
accuracy2
```

```python
cr2=classification_report(y_test,y_pred2)
print(cr2)
```

# **KNN**

# graph for different k-values

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
scoreListknn = []
for i in range(1,21):
    KNclassifier = KNeighborsClassifier(n_neighbors = i)
    KNclassifier.fit(x_train, y_train)
    scoreListknn.append(KNclassifier.score(x_test, y_test))

plt.plot(range(1,21), scoreListknn)
plt.xticks(np.arange(1,21,1))
plt.xlabel("K value")
plt.ylabel("Score")
plt.show()
KNAcc = max(scoreListknn)
print("KNN best accuracy: {:.2f}%".format(KNAcc*100))
```
